```python
import pandas as pd
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data=data.drop(columns=['RowNumber', 'CustomerId', 'Surname'], axis=1)
data
```

```python
# Convert categorical variables to numerical variables
label_encoder = LabelEncoder()
data['Gender'] = label_encoder.fit_transform(data['Gender'])
```

```python
data
```

```python
# One-hot encoding for categorical variables"geography"
geo_encoder = OneHotEncoder()
geo_encoded = geo_encoder.fit_transform(data[['Geography']]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=geo_encoder.get_feature_names_out(['Geography']))
data = pd.concat([data.drop(columns=['Geography']), geo_encoded_df], axis=1)
```

```python
data
```

```python
# independent and dependent variables
# Regression
x= data.drop(columns=['EstimatedSalary'], axis=1)
y = data['EstimatedSalary']
```

```python
x
```

```python
y
```

```python
# train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# scalling 
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
y_train
```

```python
#save  encoder and scaler
with open('geo_encoder.pkl', 'wb') as f:
    pickle.dump(geo_encoder, f)
with open('label_encoder.pkl', 'wb') as f:
    pickle.dump(label_encoder, f)
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

# ANN Regression Model Implementation

```python
# Building the model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create a Sequential model
model = Sequential(
    [
        Dense(64, activation='relu', input_shape=(x_train.shape[1],)),
        Dense(32, activation='relu'),
        Dense(1)
    ]
)

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

model.summary()
```

```python
# setup tensorboard
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime

log_dir = "regression_logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

# Early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# TRain the model
history = model.fit(x_train, y_train, epochs=100,validation_data=(x_test, y_test), callbacks=[tensorboard_callback, early_stopping])
```

```python
# Evaluate the model
loss, mae = model.evaluate(x_test, y_test)
print(f"Mean Absolute Error: {mae}")
```

```python
#make predictions
predictions = model.predict(x_test)
predictions[:5]
```
